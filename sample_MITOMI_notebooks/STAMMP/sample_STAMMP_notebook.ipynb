```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import scipy.stats
from lmfit import minimize, Parameters, report_fit
```

```python
#flag for poor GFP expression (arbitrarily determined based on examining distributions of empty chambers and positive chambers)
def eGFPFlag(df, threshold):
    df_copy = df.copy();  
    #make a column with flagging based on low eGFP
    df_copy['low_eGFP_Flag'] = df_copy['summed_button_BGsub_PostwashGFP'] < threshold
    return df_copy

# make standard curve
def DNAstdCurve(df):
    df_copy = df.copy();
    concs = df_copy['DNA_Conc']
    intensities  = df_copy['median_chamber_PrewashCy5']
    slope, intercept, r_value, p_value, std_err = scipy.stats.linregress(concs, intensities)
    df_copy['Slope_Std'] = slope
    df_copy['RSquared_Std'] = r_value**2
    df_copy['Intercept_Std'] = intercept
    df_copy['pValue_Std'] = p_value
    df_copy['stdErr_Std'] = std_err
    df_copy['Chamber_DNA_Std'] = (df_copy['median_chamber_PrewashCy5'] - intercept)/slope
    return df_copy.reset_index(drop = True)

#calibrate concentration of DNA (as calculated by nanodrop) to intensity on device to create a calibration curve
def mapConcentration(A260, concentration):
    #convert A260 into g/mL DNA
    gram_per_mL = (A260*50)*(10**(-6))
    #convert to molar concentration
    conc8_DNA = gram_per_mL * (10**3)/(660*60)
        
    if concentration == 6:
        return (conc8_DNA/1)*10**9
    elif concentration == 5:
        return (conc8_DNA/2)*10**9
    elif concentration == 4:
        return (conc8_DNA/8)*10**9
    elif concentration == 3:
        return (conc8_DNA/16)*10**9
    elif concentration == 2:
        return (conc8_DNA/32)*10**9
    else:
        return (conc8_DNA/64)*10**9

# objective function calculates residuals of fits
def objective(params, chamber_list, ratio, DNA):
    resid = np.zeros(ratio.shape)
    # make residual per data set
    for i, chamber in enumerate(chamber_list):
        resid[i-1, :] = ratio[i-1, :] - isotherm_dataset(params, chamber, DNA[i-1,:])
    # now flatten this to a 1D array, as minimize() needs
    return resid.flatten()

# function to calculate for fit
def isotherm(x,ymax,Kd):
    y = [ymax*DNA/(DNA+Kd) for DNA in x]
    return y

def isotherm_dataset(params, ch, x):
    """calc isotherm from params for data set i
    using simple, hardwired naming convention"""
    Kd = params['Kd_%s' % ch].value
    ymax = params['ymax'].value
    return isotherm(x, ymax, Kd)

def annotateExp(df):
    result = df.copy();
    
    result['Date'] = 190423;
    result['Oligo_Identity'] = 'Oligo5';
    
    return result;
```

```python
# import  prewash Cy5 data
prewash_Cy5 = pd.read_csv('STAMMP_sample_prewashcy5.csv')
prewash_Cy5 = prewash_Cy5.add_suffix('_PrewashCy5')
prewash_Cy5.rename(columns={'concentration_uM_PrewashCy5': 'step', 'x_PrewashCy5': 'x','y_PrewashCy5': 'y','id_PrewashCy5': 'id'}, inplace=True)
prewash_Cy5['Chamber_Index'] = prewash_Cy5['x'].map(str)+','+prewash_Cy5['y'].map(str)

# import postwash GFP data 
postwash_GFP = pd.read_csv('STAMMP_sample_postwashgfp.csv')
postwash_GFP = postwash_GFP.add_suffix('_PostwashGFP')
postwash_GFP.rename(columns={'step_PostwashGFP':'step','x_PostwashGFP': 'x','y_PostwashGFP': 'y','id_PostwashGFP': 'id'}, inplace=True)
postwash_GFP['Chamber_Index'] = postwash_GFP['x'].map(str)+','+postwash_GFP['y'].map(str)

#pull in postwash Cy5 data 
postwash_Cy5 = pd.read_csv('STAMMP_sample_postwashcy5.csv')
postwash_Cy5 = postwash_Cy5.add_suffix('_postwashCy5')
postwash_Cy5.rename(columns={'step_postwashCy5':'step','x_postwashCy5': 'x','y_postwashCy5': 'y','id_postwashCy5': 'id'}, inplace=True)
postwash_Cy5['Chamber_Index'] = postwash_Cy5['x'].map(str)+','+postwash_Cy5['y'].map(str)

#now perform merging: common columns are x, y, id, and step
postwash_df = pd.merge(postwash_GFP,postwash_Cy5, how='left',on=['id','step','x','y','Chamber_Index'])
rawdata_df = pd.merge(postwash_df, prewash_Cy5, how='left', on=['id','step','x','y','Chamber_Index'])

#since we want to look at the raw data in empty and full chambers, need to separate those 
rawdata_df_empty = rawdata_df[rawdata_df['id']=='Empty']
rawdata_df_notempty = rawdata_df[rawdata_df['id']!='Empty']
```

```python
#okay, so now we want to visualize the protein distribution overall in the device to assist with culling on the dataset

eGFP_test_notempty = rawdata_df_notempty[rawdata_df_notempty['step']==1]
eGFP_test_empty = rawdata_df_empty[rawdata_df_empty['step']==1]

#let's visualize the GFP expression.
g = sns.distplot(eGFP_test_notempty['summed_button_BGsub_PostwashGFP'], kde=False)
plt.title('Chambers with mutants')
plt.xticks(rotation=90)
plt.show()

g = sns.distplot(eGFP_test_empty['summed_button_BGsub_PostwashGFP'], kde=False)
plt.title('Empty Chambers')
plt.show()
```

```python
# cull data based on 99th percentile of empty chambers
threshold = 2e6;
eGFP_test_notempty = eGFPFlag(eGFP_test_notempty, threshold);

#okay, so now we need to collect the chamber IDs in the step 1 dataframe to apply to the rest of of the raw dataframe..
low_eGFP_Chambers = eGFP_test_notempty[eGFP_test_notempty['low_eGFP_Flag']==True].Chamber_Index.unique()

#now take these chambers and flag in the main dataframe 
rawdata_df['low_eGFP_Flag'] = rawdata_df['Chamber_Index'].isin(low_eGFP_Chambers)

#test to see if the filtering went properly:
flag_test = rawdata_df[rawdata_df['low_eGFP_Flag'] == False]
flag_test = flag_test[flag_test['id'] != 'Empty']
g = sns.distplot(flag_test['summed_button_BGsub_PostwashGFP'])
plt.title('Positive Expression')

#examine positional dependence on intensity

g = sns.lmplot('x','summed_button_BGsub_PostwashGFP', data = flag_test, fit_reg=False)
g.set(ylim=(0,None))

g = sns.lmplot('y','summed_button_BGsub_PostwashGFP', data = flag_test, fit_reg=False)
g.set(ylim=(0,None))
```

```python
#okay, so now we've flagged the chambers. What we need to do now is create per-chamber calibration curves.

#first, start with mapping each step of the process to a DNA concentration.
rawdata_df['DNA_Conc'] = rawdata_df['step'].apply(lambda x: mapConcentration(2.6, x)) # 2.6 value is A260 of highest concentration sample

#okay, now bin the raw dataframe by Chamber_Index and apply the DNA standard curve on a chamber-by-chamber basis.
rawdata_df_grouped = rawdata_df.groupby('Chamber_Index')
rawdata_df = rawdata_df_grouped.apply(DNAstdCurve)
```

```python
#Let me look at the DNA calibration curves across muts.

flag_test = rawdata_df[rawdata_df['low_eGFP_Flag'] == False]
g = sns.FacetGrid(flag_test, col = "id",col_wrap=28, height = 4, hue='y')
g = g.map(plt.scatter,"Chamber_DNA_Std", "median_chamber_PrewashCy5")
plt.legend()
plt.show()

# let me see how consistent/good the curves are

g = sns.distplot(flag_test['Slope_Std'], kde=False)

g = sns.distplot(flag_test['RSquared_Std'],kde=False)
```

```python
#let's see the slope as function of position on the device.

g = sns.lmplot('x','Slope_Std',data=flag_test, fit_reg = False)
g = sns.lmplot('y','Slope_Std',data=flag_test, fit_reg = False)
```

```python
rawdata_df['R_bSub'] = rawdata_df['summed_button_BGsub_postwashCy5']/rawdata_df['summed_button_BGsub_PostwashGFP']

rawdata_df = annotateExp(rawdata_df)

filtered_df = rawdata_df[rawdata_df['low_eGFP_Flag'] == False]
filtered_df = filtered_df[filtered_df['id'] != 'Empty']
filtered_df = filtered_df[filtered_df['R_bSub'] > 0]
filtered_df = filtered_df[filtered_df['Chamber_DNA_Std'] > 0]
```

```python
# plot data before curve fits

g = sns.FacetGrid(filtered_df, col = "id",col_wrap=15, height = 4,hue = "y")
g = g.map(plt.scatter, "Chamber_DNA_Std","R_bSub")
plt.ylim(0,1.5);
```

```python
# create new dataframe with only the data necessary to fit curves
GFIT_Input = pd.DataFrame();

GFIT_Input['chamber'] = filtered_df['Chamber_Index'].apply(lambda x: x.replace(',','_'))
GFIT_Input['id'] = filtered_df['id']
GFIT_Input['R'] = filtered_df['R_bSub']
GFIT_Input['DNA'] = filtered_df['Chamber_DNA_Std']
GFIT_Input = GFIT_Input.dropna().reset_index(drop=True)

# export for safekeeping
GFIT_Input.to_csv('sample_GFIT_input.csv', sep=',', index = False)
```

```python
# SET UP FIT PARAMETERS

fit_params = Parameters()

for i in GFIT_Input['chamber'].unique():
    fit_params.add( 'Kd_%s' % i, value=1e3, min=1e1,  max=1e5)

Rmax_guess = np.max(GFIT_Input['R'])
fit_params.add('ymax', value=Rmax_guess, min=0.1,  max=5)
```

```python
chamber_list = list(set(GFIT_Input['chamber']))
n_chambers = len(chamber_list)

max_len = np.max(GFIT_Input['chamber'].value_counts())
data_ratios = np.zeros((n_chambers, max_len))
data_chambers = np.zeros((n_chambers, max_len))

for j, chamber in enumerate(chamber_list):
    ratios = list(GFIT_Input[GFIT_Input['chamber']==chamber]['R'])
    data_ratios[int(j)-1,:len(ratios)] = ratios
    DNA = list(GFIT_Input[GFIT_Input['chamber']==chamber]['DNA'])
    data_chambers[int(j)-1,:len(DNA)] = DNA


result_global = minimize(objective, fit_params, args=(chamber_list, data_ratios, data_chambers))  

Kd_fits = [result_global.params['Kd_%s' % k].value for k in chamber_list]
rmax_fit = result_global.params['ymax'].value

print(rmax_fit)
```

```python
# make dataframe with Kds and export

Kd_df = pd.DataFrame(np.array([Kd_fits, chamber_list]).T, columns=['Kd', 'chamber'])
Kd_df['Kd'] = Kd_df['Kd'].astype(float)
Kd_df['Chamber_Index'] = Kd_df['chamber'].apply(lambda x: x.replace('_',','))

GFIT_Input.merge(Kd_df, on='chamber').to_csv('sample_GFIT_output.csv', sep=',', index = False)
filtered_df = filtered_df.reset_index(drop=True).merge(Kd_df, on='Chamber_Index')
```

## plot binding curves

```python
mut_order = np.load('mutant_order.npy')

n_muts = len(GFIT_Input['id'].unique())
cols = 8
rows = int(n_muts/cols)+1

x_array = np.linspace(0,3500,20)

fig = plt.figure(figsize=(cols*5,rows*5))
panel_count = 0
for i, mut in enumerate(mut_order):
    data_by_mut = GFIT_Input[GFIT_Input['id']==mut]
    if len(data_by_mut) > 0:
        panel_count += 1
        ax = fig.add_subplot(rows,cols,panel_count)
        mean_Kd = np.mean(filtered_df[filtered_df['id']==mut]['Kd'].unique())
        stdev_Kd = np.std(filtered_df[filtered_df['id']==mut]['Kd'].unique())
        ax.scatter(data_by_mut['DNA'],data_by_mut['R'])
        ax.set_title(mut, fontsize=20)
        for chamber in data_by_mut['chamber'].unique():
            Kd = list(Kd_df[Kd_df['chamber']==chamber]['Kd'])[0]
            ax.plot(x_array, isotherm(x_array, rmax_fit, Kd), c='gray',ls='--')
        ax.text(0,1.8,'Kd = %.f $\pm$ %.f nM' % (mean_Kd, stdev_Kd), fontsize=14)
        ax.set_ylim(0,2.)
        if panel_count % cols == 1:
            ax.set_ylabel('DNA/TF ratio', fontsize=16)
        if panel_count > n_muts - cols:
            ax.set_xlabel('[DNA] (nM)', fontsize=16)

fig.tight_layout()
fig.savefig('binding_curves.pdf')
```
