```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from scipy import stats
import pickle

np.random.seed(0)
```

## are repeats particularly enriched in broadly active enhancers?

what is repeat enrichment as a function of # of cell types?

```python
# load numpy arrays containing enrichment scores (as odds ratios) for enhancers as a function of # cell types in which they are active
OR_list = np.load('celltypes/celltypes_OR.npy')
OR_err = np.load('celltypes/celltypes_ORerr.npy')

# calculate 95% confidence intervals
upper = OR_list*np.exp(1.96*np.array(OR_err))-OR_list
lower = OR_list-OR_list/np.exp(1.96*np.array(OR_err))
```

```python
# load dictionaries containing enrichment scores (as odds ratios) for shuffled enhancer lists 
with open('celltypes/celltypes_OR_rand.pkl','rb') as f:
    OR_rand = pickle.load(f)
    
with open('celltypes/celltypes_ORerr_rand.pkl','rb') as f:
    OR_rand_err = pickle.load(f)
```

```python
fig = plt.figure(figsize=(10,5))
ax = fig.add_subplot(111)

thresholds = np.arange(0,231,10)

# plot enrichment for the genome
ax.errorbar(thresholds, OR_list,[lower,upper],
            marker='o', capsize=4, c='r')

# plot enrichment for shuffled enhancer list
y = [np.median(OR_rand[j]) for j in OR_rand]
yerr = [np.std(OR_rand[j]) for j in OR_rand]
ax.errorbar(thresholds, y, 1.96*np.array(yerr)/10,
            marker='o', capsize=4, c='gray') #error bars are 95% CI

ax.set_ylabel('Enrichment of repeats in enhancers\n(odds ratio)', fontsize=16)
ax.set_xlabel('Threshold # of cell types', fontsize=16)

rho_genome = stats.spearmanr(thresholds, OR_list)[0]
rho_shuffled = stats.spearmanr(thresholds, y)[0]
ax.tick_params(labelsize=14)

ax.legend(['Actual genome: $R^2_s=$ %.2f' % rho_genome**2,
           'Shuffled labels: $R^2_s=$ %.2f' % rho_shuffled**2], fontsize=14)

fig.tight_layout()
# fig.savefig('enrichment_vs_numbercelltypes.png')
```

# Code for generating arrays

```python
from collections import defaultdict
from tqdm import tqdm
```

```python
def enrichment_calc(enhancer_dict, trf_dict, seqlen):
    '''
    Calculate enrichment of STRs in enhancers.
    
    1. construct contingency table:
              
            | enhancer | not enhancer |
    -----------------------------------
    STR     |    q1    |      q2      |
    not STR |    q3    |      q4      |
    -----------------------------------
    
    2. calculate OR = q1*q4/(q2*q4)
    3. calculate std err = √(1/q1+1/q2+1/q3+1/q4)
    '''
    q1 = np.sum([len(np.intersect1d(enhancer_dict[x], trf_dict[x])) for x in enhancer_dict])
    q2 = np.sum([len(enhancer_dict[x]) - len(np.intersect1d(enhancer_dict[x], trf_dict[x])) for x in enhancer_dict])
    q3 = np.sum([len(trf_dict[x]) - len(np.intersect1d(enhancer_dict[x], trf_dict[x])) for x in enhancer_dict])
    q4 = seqlen - np.sum([q1, q2, q3])
    
    # calculate odds ratio
    OR = q1*q4/(q2*q3)
    # calculate standard error of OR
    stderr = np.sqrt((1/q1+1/q2+1/q3+1/q4))

    return OR, stderr
```

```python
# load some basic info about human genome
chromsize_df = pd.read_csv('hg38.chrom.sizes.txt', delimiter='\t', names=['size']).drop('chrY')
chroms = list(chromsize_df.index)
genomesize = np.sum(chromsize_df.loc[chroms])[0]

# load Enhancer Atlas list of enhancers
human_enhancers = pd.read_csv('/Users/connor/Documents/fordycelab/TRF/enhancer_lists/enhancer_atlas_2/hg38_enhancers.csv')
```

## Load in data from TRF

```python
# load in list of STRs in the genome
trf_dfs = {}
for chrom in chroms:
    trf_dfs[chrom] = pd.read_csv('/Users/connor/Documents/fordycelab/TRF/hg38/TRF_output/%s_repeats.csv' % chrom)
    
# organize STRs into dictionary with a list of genomic coordinates for each chromosome
trf_dict = {}
for chrom in tqdm(trf_dfs):
    trf_dfs[chrom] = trf_dfs[chrom][trf_dfs[chrom]['consensus_size']<5]
    trf_dict[chrom] = np.hstack([np.arange(row['start'], row['end']+1) for i,row in trf_dfs[chrom].iterrows()])
    
trf_dict = {k: np.unique(v) for k,v in trf_dict.items()}
```

```python
# specify a moving threshold and calculate repeat enrichment for enhancers with more cell types than the threshold
thresholds = np.arange(0,255,5)
OR_list = []
OR_err = []

# for each threshold x, calculate enrichment of STRs in enhancers active in x cells
for thresh in tqdm(thresholds):
    temp_df = human_enhancers[human_enhancers['num_celltypes']>thresh]
    temp_dict = {}
    for i in chroms:
        try:
            temp_dict[i] = np.hstack([np.arange(row['start'], row['end']+1) for i,row in temp_df[temp_df['chr']==i].iterrows()])
        except:
            pass
    temp_dict = {k: np.unique(v) for k,v in temp_dict.items()}
    OR, stderr = enrichment_calc(temp_dict, trf_dict, np.sum(chromsize_df.loc[chroms])[0])
    OR_list.append(OR)
    OR_err.append(stderr)
    
# save as numpy arrays
np.save('celltypes/celltypes_OR.npy',OR_list)
np.save('celltypes/celltypes_ORerr.npy',OR_err)
```

```python
thresholds = np.arange(0,255,10)

EA_len = len(human_enhancers)
OR_rand = defaultdict(list)
OR_rand_err = defaultdict(list)

N = 100 # number of iterations

# for each threshold activity value, sample n enhancers and compute STR enrichment
for thresh in tqdm(thresholds):
    n = np.count_nonzero(human_enhancers['num_celltypes']>thresh) 
    for i in range(N):
        rand_locs = np.random.choice(EA_len, size=n, replace=False) # select n enhancers 
        temp_df = human_enhancers.loc[rand_locs]
        temp_dict = {}
        for chrom in chroms:
            try:
                temp_dict[chrom] = np.hstack([np.arange(row['start'], row['end']+1) for i,row in temp_df[temp_df['chr']==chrom].iterrows()])
            except:
                pass
        temp_dict = {k: np.unique(v) for k,v in temp_dict.items()}
        OR, stderr = enrichment_calc(temp_dict, trf_dict, genomesize)
        OR_rand[thresh].append(OR)
        OR_rand_err[thresh].append(stderr)

# save as pickles
with open('celltypes/celltypes_OR_rand.pkl','wb') as f:
    pickle.dump(OR_rand,f)
with open('celltypes/celltypes_ORerr_rand.pkl','wb') as f:
    pickle.dump(OR_rand_err,f) 
```
