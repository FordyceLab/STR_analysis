```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from scipy import stats
from tqdm import tqdm
from itertools import combinations, product
import seaborn as sns
from Bio.Seq import Seq
from Levenshtein import distance
import pickle
```

```python
# load data to be analyzed
all_Z = pd.read_csv('../../uniprobe_data/Zscores_byTF.csv')
repeat_df = pd.read_csv('../../uniprobe_data/Zscores_repeatsonly_byTF.csv')
info = pd.read_csv('../../uniprobe_data/CISBP_PBM_info_families.csv')
```

## Levenshtein distance
https://github.com/ztane/python-Levenshtein/

for each repeat type, get the minimum levenshtein distance of each of the PWM positions and reverse complement

```python
def get_minimum_LD(motif, repeat):
    temp = repeat*2
    all_frames = [temp[j:j+8] for j in range(len(temp))]
    all_frames.extend([str(Seq(temp[j:j+8]).reverse_complement()) for j in range(len(temp))])
    LD = [distance(motif, x) for x in all_frames]
    return np.min(LD)

def get_motif(motif_id):
    # motifs can be downloaded from CISBP
    file = 'pwms/%s.txt' % motif_id
    pwm = pd.read_csv(file, index_col=0, delimiter='\t')
    norm = pwm.T/pwm.max(axis=1)
    temp_dict = norm[norm>0.6].to_dict()
    pos = []
    for idx in temp_dict:
        pos.append([key for key in temp_dict[idx] if not np.isnan(temp_dict[idx][key])])
    return [''.join(p) for p in product(*pos)]   

def get_minimum_LD_allmotifs(motifs, repeat):
    temp = repeat*3
    LD = []
    for motif in motifs:
        motif_len = len(motif)
        all_frames = [temp[j:j+motif_len] for j in range(len(temp)-motif_len+1)]
        all_frames.extend([str(Seq(temp[j:j+motif_len]).reverse_complement()) for j in range(len(temp)-motif_len+1)])
        LD.extend([distance(motif, x) for x in all_frames])
    return np.min(LD)
```

```python
# to repeat the analysis without re-calculating the levenshtein distances, you can load this dictionary
with open('repeat_LD.pkl','rb') as f:
    repeat_LD = pickle.load(f)
```

```python
# calculate spearman correlations between motif similarity and binding. How similar are favored repeats to the binding motif?
spearman_LD_correlations = {}

for TF in tqdm(repeat_LD):
    spearman_LD_correlations[TF] = stats.spearmanr(repeat_LD[TF], repeat_df[TF])[0]
    
families = [list(info[info['col_name']==x]['Family_Name_Condensed'])[0] for x in list(spearman_LD_correlations.keys())]

spearman_df = pd.DataFrame(data={'TF':list(spearman_LD_correlations.keys()),
                                 'spearman':list(spearman_LD_correlations.values()),
                                 'family':families})
```

```python
fig = plt.figure(figsize=(8,3))

sns.set_theme(style="white", rc={"axes.facecolor": (0, 0, 0, 0)})


order = ['AT hook', 'ARID/BRIGHT', 'POU', 'Sox', 'Homeodomain', 'Forkhead', 'GATA',
         'Myb/SANT', 'bZIP', 'Nuclear receptor', 'bHLH', 'AP2', 'WRKY', 'ZnF',
         'T-box', 'Ets', 'E2F','Other']

pal = sns.cubehelix_palette(len(np.unique(spearman_df[['family']])), rot=-.2, light=.8)
g = sns.FacetGrid(spearman_df, row="family", hue="family", height=0.8,
                  aspect=12, palette=pal, row_order=order, hue_order=order)



# Draw the densities in a few steps
g.map(sns.kdeplot, "spearman", bw_adjust=.5, clip_on=False, fill=True, alpha=1, linewidth=1.5)
g.map(sns.kdeplot, "spearman", clip_on=False, color="k", lw=1, bw_adjust=.5)
g.map(plt.axhline, y=0, lw=2, c='k', clip_on=False)
# g.map(plt.axhline, y=-0.04, lw=1, c='k', clip_on=True)
g.map(plt.axvline, x=0, lw=1, c='k', clip_on=True)


def label(x, color, label):
    ax = plt.gca()
    ax.text(1, .3, label, fontweight="bold", color=color, ha="right",
            va="center", transform=ax.transAxes, fontsize=20)

g.map(label, "family") 


# Set the subplots to overlap
g.fig.subplots_adjust(hspace=-0.5)

# Remove axes details that don't play well with overlap
g.set(ylabel="")
g.set_titles("")
g.set(yticks=[])
g.set(xlabel="spearman correlation (r)")
g.despine(bottom=False, left=True)

means = [spearman_df.groupby('family').mean().loc[x,'spearman'] for x in order]
for i, ax in enumerate(g.axes.flat):
    kdeline = ax.lines[0]
    xs = kdeline.get_xdata()
    ys = kdeline.get_ydata()
    height = np.interp(means[i], xs, ys)/6
    ax.axvline(means[i], 0, height, color='k', ls='-')

fig.tight_layout()

# g.savefig('motif_repeat_correlation_bydomain.png')
# g.savefig('motif_repeat_correlation_bydomain.eps')
```

```python
fig = plt.figure(figsize=(6,4.5))

TF = 'PHO4 (Saccharomyces cerevisiae)'

ax = fig.add_subplot(111)

x = repeat_LD[TF] + list(range(8))
y = list(repeat_df[TF]) + [-10]*8
sns.swarmplot(x=x, y=y, color='gray')

# ax.set_xticklabels(x)
ax.set_ylabel('Repeat Z-score', fontsize=18)
ax.set_xlabel('Levenshtein distance', fontsize=18)
ax.tick_params(labelsize=14)
ax.text(0,38,"$R_s = $ %.3f" % stats.spearmanr(repeat_LD[TF], repeat_df[TF])[0], fontsize=14)
ax.set_ylim(-7,42)

fig.tight_layout()
# fig.savefig('Z_vs_levenshtein_%s.eps' % TF.split(' ')[0])
```
